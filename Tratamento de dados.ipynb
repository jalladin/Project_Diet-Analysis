### "The dataset aims to analyze how the variables interact and influence weight fluctuations, providing a valuable resource for researchers and practitioners in nutrition and health"

```python
#Import data

dataset = pd.read_csv("dataset/weight_change_dataset.csv", sep= ",")

dataset
```

- **Quantitative collumns:**
    - Continuos (float): 
        - Age, 
        - Current Weight (lbs), 
        - BMR (Calories), 
        - Daily Calories Consumed, 
        - Daily Caloric Surplus/Deficit, 
        - Weight Change (lbs), 
        - Stress Level, 
        - Final Weight (lbs)
    - Discrete (int): 
        - Participant ID, 
        - Duration (weeks)


- **Qualitative collumns:**
    - Gender, (Ok)
    - Physical Activity Level, (OK) 
    - Sleep Quality (OK)

**Columns meaning:**
 - BMR = Minimum calories that the body needs to consum daily\
 - Daily Caloric Surplus = means that the person consumed more calories than they expended.\
 - Daily Caloric Deficit = occurs when a person consumes fewer calories than the body uses

The column "Daily Caloric" is the rest of "Daily Calories Consumed" from BMR (Calories). The result can be surplus or deficit.
